```python
# Calculator :
a = int(input("Enter first number > "))
b = int(input("enter the second number > "))
oper = input("Enter the opertion ( + - * / ) > ")

if oper=="+":
  print("Ans : ",a+b)

elif oper=="-":
  print("Ans > ",a-b)

elif oper=="*":
  print("Ans > ",a*b)

elif(oper=="/"):
  print("Ans > ",a/b)
```
