```python
# Problem :
"""Make a average marks calculator that takes input from user and do the average of marks."""
# Solution :

s1 = int(input("Enter marks for subject 1 > "))
s2 = int(input("Enter marks for subject 2 > "))
s3 = int(input("Enter marks for subject 3 > "))
s4 = int(input("Enter marks for subject 4 > "))
s5 = int(input("Enter marks for subject 5 > "))

print(f"The total marks obtained is {s1+s2+s3+s4+s5}/500")
print(f"The total average obtained is {(s1+s2+s3+s4+s5)/5}/100")
```

```python
# Problem :
"""Write a code to calculate the area of circle and rectangle by using their formulas."""
# Solution :

Radius = eval(input ("Enter the radius of circle : ")) # gettign radius as input
print (f"The area of circle is : {3.14*Radius*Radius}") # calculation and displaying

Length = eval(input ("Enter the Length : ")) # gettign Length as input
Height = eval(input ("Enter the Height : ")) # gettign Height as input
print (f"The area of Rectangle is : {Length*Height}") # calculation and displaying
```
